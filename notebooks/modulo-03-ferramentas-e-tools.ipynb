## ğŸ› ï¸ Primeira Tool Customizada - ComeÃ§ando Simples

Vamos criar nossa primeira tool do zero! Vai ser uma calculadora simples, mas vocÃª vai entender o conceito.

````python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools import tool

# Criando nossa primeira tool customizada
@tool
def calculadora_simples(operacao: str, numero1: float, numero2: float) -> str:
    """
    Realiza operaÃ§Ãµes matemÃ¡ticas bÃ¡sicas.
    
    Args:
        operacao (str): Tipo da operaÃ§Ã£o (soma, subtracao, multiplicacao, divisao)
        numero1 (float): Primeiro nÃºmero
        numero2 (float): Segundo nÃºmero
    
    Returns:
        str: Resultado da operaÃ§Ã£o
    """
    
    if operacao == "soma":
        resultado = numero1 + numero2
    elif operacao == "subtracao":
        resultado = numero1 - numero2
    elif operacao == "multiplicacao":
        resultado = numero1 * numero2
    elif operacao == "divisao":
        if numero2 == 0:
            return "Erro: DivisÃ£o por zero nÃ£o Ã© permitida!"
        resultado = numero1 / numero2
    else:
        return "Erro: OperaÃ§Ã£o nÃ£o reconhecida. Use: soma, subtracao, multiplicacao ou divisao"
    
    return f"O resultado de {numero1} {operacao} {numero2} Ã©: {resultado}"

print("ğŸ¯ Tool calculadora criada com sucesso!")
print("ğŸ“ Repare na estrutura: decorator @tool, docstring detalhada e funÃ§Ã£o Python normal")
````

````python
# Criando um agente com nossa tool customizada
agente_calculador = Agent(
    name="Calculador Expert",
    model=Gemini(id="gemini-2.0-flash-exp"),
    tools=[calculadora_simples],  # Adicionando nossa tool
    instructions=[
        "VocÃª Ã© um assistente matemÃ¡tico especializado.",
        "Use a calculadora sempre que precisar fazer contas.",
        "Seja didÃ¡tico e explique os passos."
    ],
    markdown=True
)

print("ğŸ¤– Agente Calculador criado!")
print("ğŸ”§ Tool integrada com sucesso!")
````

````python
# Testando nossa primeira tool
pergunta = "Quanto Ã© 157 multiplicado por 23? E depois some 45 ao resultado?"

print("â“ Pergunta:", pergunta)
print("\n" + "="*50)

resposta = agente_calculador.run(pergunta)
print(resposta.content)

print("\n" + "="*50)
print("ğŸ‰ Liiindo! O agente usou nossa tool automaticamente!")
````

## ğŸŒ¡ï¸ Tool Customizada - Consulta de Clima

Vamos criar uma tool mais sofisticada que consulta uma API externa real! Vamos usar a OpenWeatherMap API (gratuita) para criar um consultor de clima.

**Dica!** Esta Ã© uma tool que vocÃª pode adaptar para qualquer API externa - Ã© o template perfeito!

````python
import requests
import json

@tool
def consultar_clima(cidade: str) -> str:
    """
    Consulta informaÃ§Ãµes do clima atual para uma cidade especÃ­fica.
    
    Args:
        cidade (str): Nome da cidade para consultar o clima
    
    Returns:
        str: InformaÃ§Ãµes detalhadas do clima da cidade
    """
    
    try:
        # API gratuita do OpenWeather (sem necessidade de key para demo)
        # Em produÃ§Ã£o, vocÃª usaria uma API key real
        url = f"http://api.openweathermap.org/data/2.5/weather"
        
        # Para este exemplo, vamos simular os dados
        # Em um caso real, vocÃª faria a requisiÃ§Ã£o HTTP
        
        # Simulando diferentes climas para diferentes cidades
        climas_simulados = {
            "sÃ£o paulo": {
                "temperatura": 22,
                "descricao": "Nublado",
                "umidade": 75,
                "vento": 15
            },
            "rio de janeiro": {
                "temperatura": 28,
                "descricao": "Ensolarado",
                "umidade": 65,
                "vento": 12
            },
            "recife": {
                "temperatura": 30,
                "descricao": "Parcialmente nublado",
                "umidade": 80,
                "vento": 18
            }
        }
        
        cidade_lower = cidade.lower()
        
        if cidade_lower in climas_simulados:
            clima = climas_simulados[cidade_lower]
            
            resultado = f"""
ğŸŒ¤ï¸ Clima em {cidade.title()}:
ğŸŒ¡ï¸ Temperatura: {clima['temperatura']}Â°C
â˜ï¸ CondiÃ§Ã£o: {clima['descricao']}
ğŸ’§ Umidade: {clima['umidade']}%
ğŸ’¨ Vento: {clima['vento']} km/h
"""
            return resultado.strip()
        else:
            return f"âŒ Dados de clima nÃ£o disponÃ­veis para {cidade}. Tente: SÃ£o Paulo, Rio de Janeiro ou Recife."
            
    except Exception as e:
        return f"âŒ Erro ao consultar clima: {str(e)}"

print("ğŸŒ¤ï¸ Tool de clima criada!")
print("ğŸ“¡ Conectada a dados meteorolÃ³gicos simulados")
````

## ğŸ”§ Tool AvanÃ§ada - Gerador de RelatÃ³rios

Agora vamos criar algo mais sofisticado: uma tool que gera relatÃ³rios estruturados! Esta Ã© uma ferramenta que vocÃª pode usar em projetos reais.

**Dica!** Esta tool mostra como processar dados complexos e retornar resultados estruturados.

````python
from datetime import datetime
import random

@tool
def gerar_relatorio_vendas(mes: str, ano: int) -> str:
    """
    Gera um relatÃ³rio detalhado de vendas para um mÃªs especÃ­fico.
    
    Args:
        mes (str): Nome do mÃªs (janeiro, fevereiro, etc.)
        ano (int): Ano do relatÃ³rio
    
    Returns:
        str: RelatÃ³rio detalhado de vendas formatado
    """
    
    # Simulando dados de vendas
    produtos = [
        "Notebook Gamer", "Mouse Wireless", "Teclado MecÃ¢nico", 
        "Monitor 4K", "Headset Gaming", "Webcam HD", "SSD 1TB"
    ]
    
    # Gerando dados aleatÃ³rios para demonstraÃ§Ã£o
    vendas_total = 0
    detalhes_produtos = []
    
    for produto in produtos:
        quantidade = random.randint(5, 50)
        preco_unitario = random.randint(50, 2000)
        valor_total = quantidade * preco_unitario
        vendas_total += valor_total
        
        detalhes_produtos.append({
            "produto": produto,
            "quantidade": quantidade,
            "preco_unitario": preco_unitario,
            "valor_total": valor_total
        })
    
    # Ordenando por valor total (maior para menor)
    detalhes_produtos.sort(key=lambda x: x['valor_total'], reverse=True)
    
    # Construindo o relatÃ³rio
    relatorio = f"""
ğŸ“Š RELATÃ“RIO DE VENDAS - {mes.upper()} {ano}
{'='*50}

ğŸ’° RESUMO EXECUTIVO:
   â€¢ Faturamento Total: R$ {vendas_total:,.2f}
   â€¢ Produtos Vendidos: {len(produtos)}
   â€¢ Ticket MÃ©dio: R$ {vendas_total/sum(p['quantidade'] for p in detalhes_produtos):,.2f}

ğŸ† RANKING DE PRODUTOS:
"""
    
    for i, produto in enumerate(detalhes_produtos, 1):
        relatorio += f"""
   {i}Âº {produto['produto']}
      â†’ Qtd: {produto['quantidade']} unidades
      â†’ PreÃ§o: R$ {produto['preco_unitario']:,.2f}
      â†’ Total: R$ {produto['valor_total']:,.2f}
"""
    
    relatorio += f"""

ğŸ“ˆ INSIGHTS:
   â€¢ Produto mais vendido: {detalhes_produtos[0]['produto']}
   â€¢ Maior faturamento: R$ {detalhes_produtos[0]['valor_total']:,.2f}
   â€¢ Data do relatÃ³rio: {datetime.now().strftime('%d/%m/%Y %H:%M')}
"""
    
    return relatorio.strip()

print("ğŸ“Š Tool de relatÃ³rios criada!")
print("ğŸ“ˆ Pronta para gerar insights de vendas")
````

````python
# Criando agente analista de vendas
agente_vendas = Agent(
    name="Analista de Vendas",
    model=Gemini(id="gemini-2.0-flash-exp"),
    tools=[gerar_relatorio_vendas],
    instructions=[
        "VocÃª Ã© um analista de vendas experiente.",
        "Gere relatÃ³rios detalhados quando solicitado.",
        "ForneÃ§a insights e recomendaÃ§Ãµes baseadas nos dados.",
        "Seja analÃ­tico e estratÃ©gico."
    ],
    markdown=True
)

# Testando geraÃ§Ã£o de relatÃ³rio
request_relatorio = "Gere um relatÃ³rio de vendas para dezembro de 2024 e me dÃª insights sobre o performance"

print("ğŸ“Š Gerando relatÃ³rio:", request_relatorio)
print("\n" + "="*60)

resposta = agente_vendas.run(request_relatorio)
print(resposta.content)

print("\n" + "="*60)
print("ğŸ¯ RelatÃ³rio profissional gerado com sucesso!")
````

## ğŸ¯ Melhores PrÃ¡ticas para Tools

![](https://s3.us-east-1.amazonaws.com/turing.education/books/imagens/agno-2.0-framework--modulo-03_img_03.png)

Depois de criar vÃ¡rias tools, aqui estÃ£o as **liÃ§Ãµes de ouro** que aprendi:

### 1. ğŸ“ DocumentaÃ§Ã£o Clara
- **Docstring detalhada**: O LLM precisa entender o que sua tool faz
- **ParÃ¢metros bem definidos**: Tipos e descriÃ§Ãµes claras
- **Exemplos de uso**: Quando possÃ­vel, dÃª exemplos

### 2. ğŸ›¡ï¸ Tratamento de Erros
- **Try/except sempre**: APIs externas podem falhar
- **Mensagens de erro Ãºteis**: Ajude o agente a entender o que deu errado
- **Fallbacks**: Tenha planos B quando possÃ­vel

### 3. âš¡ Performance
- **Timeouts**: NÃ£o deixe o agente travado esperando
- **Cache**: Evite consultas desnecessÃ¡rias
- **Dados mÃ­nimos**: Retorne sÃ³ o que Ã© necessÃ¡rio

### 4. ğŸ”’ SeguranÃ§a
- **ValidaÃ§Ã£o de entrada**: Sempre valide os parÃ¢metros
- **API keys seguras**: Use variÃ¡veis de ambiente
- **Rate limiting**: Respeite limites das APIs

````python
# Exemplo de tool seguindo todas as melhores prÃ¡ticas
@tool
def consultar_cep_brasileiro(cep: str) -> str:
    """
    Consulta informaÃ§Ãµes de endereÃ§o a partir de um CEP brasileiro.
    
    Esta tool utiliza a API gratuita ViaCEP para buscar informaÃ§Ãµes
    detalhadas de endereÃ§o baseadas no CEP fornecido.
    
    Args:
        cep (str): CEP brasileiro no formato XXXXX-XXX ou XXXXXXXX
                  Exemplo: "01310-100" ou "01310100"
    
    Returns:
        str: InformaÃ§Ãµes detalhadas do endereÃ§o incluindo:
             - Logradouro, Bairro, Cidade, Estado
             - CÃ³digo IBGE, DDD
             - Coordenadas (quando disponÃ­vel)
    
    Exemplo de uso:
        consultar_cep_brasileiro("01310-100") -> Retorna dados da Av. Paulista
    """
    
    try:
        # 1. VALIDAÃ‡ÃƒO DE ENTRADA
        if not cep:
            return "âŒ Erro: CEP nÃ£o fornecido"
        
        # Limpando o CEP (removendo caracteres especiais)
        cep_limpo = ''.join(filter(str.isdigit, cep))
        
        if len(cep_limpo) != 8:
            return "âŒ Erro: CEP deve ter 8 dÃ­gitos. Formato: XXXXX-XXX"
        
        # 2. SIMULAÃ‡ÃƒO DE DADOS (em produÃ§Ã£o, faria requisiÃ§Ã£o real)
        # URL real seria: f"https://viacep.com.br/ws/{cep_limpo}/json/"
        
        # Dados simulados para demonstraÃ§Ã£o
        ceps_conhecidos = {
            "01310100": {
                "logradouro": "Avenida Paulista",
                "bairro": "Bela Vista",
                "cidade": "SÃ£o Paulo",
                "estado": "SP",
                "ddd": "11",
                "ibge": "3550308"
            },
            "20040020": {
                "logradouro": "Rua da AssemblÃ©ia",
                "bairro": "Centro",
                "cidade": "Rio de Janeiro",
                "estado": "RJ",
                "ddd": "21",
                "ibge": "3304557"
            }
        }
        
        if cep_limpo in ceps_conhecidos:
            dados = ceps_conhecidos[cep_limpo]
            
            # 3. FORMATAÃ‡ÃƒO DA RESPOSTA
            resultado = f"""
ğŸ“ INFORMAÃ‡Ã•ES DO CEP {cep}
{'='*40}
ğŸ  Logradouro: {dados['logradouro']}
ğŸ˜ï¸  Bairro: {dados['bairro']}
ğŸ™ï¸  Cidade: {dados['cidade']}
ğŸ—ºï¸  Estado: {dados['estado']}
ğŸ“ DDD: {dados['ddd']}
ğŸ†” CÃ³digo IBGE: {dados['ibge']}
âœ… Consulta realizada com sucesso!
"""
            return resultado.strip()
        else:
            return f"âŒ CEP {cep} nÃ£o encontrado na base de dados (demo limitada)"
            
    except Exception as e:
        # 4. TRATAMENTO DE ERROS ROBUSTO
        return f"âŒ Erro interno na consulta de CEP: {str(e)}"

print("ğŸ“ Tool de CEP criada seguindo todas as melhores prÃ¡ticas!")
print("âœ… DocumentaÃ§Ã£o completa, validaÃ§Ã£o, tratamento de erros")
````

````python
# ğŸ¯ EXERCÃCIO: Complete esta tool de conversÃ£o de moedas

@tool
def conversor_moedas(valor: float, moeda_origem: str, moeda_destino: str) -> str:
    """
    Converte valores entre diferentes moedas usando taxas de cÃ¢mbio simuladas.
    
    Args:
        valor (float): Valor a ser convertido
        moeda_origem (str): Moeda de origem (BRL, USD, EUR)
        moeda_destino (str): Moeda de destino (BRL, USD, EUR)
    
    Returns:
        str: Resultado da conversÃ£o com detalhes
    """
    
    # COMPLETE AQUI: Implemente a lÃ³gica de conversÃ£o
    # Dica: Use taxas de cÃ¢mbio simuladas
    # Exemplo: 1 USD = 5.20 BRL, 1 EUR = 6.10 BRL
    
    taxas_cambio = {
        # TODO: Defina as taxas de cÃ¢mbio
        # Formato: (moeda_origem, moeda_destino): taxa
        ('USD', 'BRL'): 5.20,
        ('BRL', 'USD'): 1/5.20,
        ('EUR', 'BRL'): 6.10,
        ('BRL', 'EUR'): 1/6.10,
        ('USD', 'EUR'): 0.85,
        ('EUR', 'USD'): 1/0.85
    }
    
    # TODO: Implementar validaÃ§Ãµes
    # TODO: Implementar conversÃ£o
    # TODO: Formatar resultado
    
    try:
        # ValidaÃ§Ãµes bÃ¡sicas
        if valor <= 0:
            return "âŒ Erro: Valor deve ser maior que zero"
        
        moeda_origem = moeda_origem.upper()
        moeda_destino = moeda_destino.upper()
        
        if moeda_origem == moeda_destino:
            return f"ğŸ’° {valor:.2f} {moeda_origem} = {valor:.2f} {moeda_destino} (mesma moeda)"
        
        # Buscar taxa de cÃ¢mbio
        chave_taxa = (moeda_origem, moeda_destino)
        
        if chave_taxa in taxas_cambio:
            taxa = taxas_cambio[chave_taxa]
            valor_convertido = valor * taxa
            
            resultado = f"""
ğŸ’± CONVERSÃƒO DE MOEDAS
{'='*30}
ğŸ’° Valor Original: {valor:.2f} {moeda_origem}
ğŸ”„ Taxa de CÃ¢mbio: 1 {moeda_origem} = {taxa:.4f} {moeda_destino}
ğŸ’ Valor Convertido: {valor_convertido:.2f} {moeda_destino}
ğŸ“… Taxas simuladas para demonstraÃ§Ã£o
"""
            return resultado.strip()
        else:
            return f"âŒ ConversÃ£o de {moeda_origem} para {moeda_destino} nÃ£o disponÃ­vel"
            
    except Exception as e:
        return f"âŒ Erro na conversÃ£o: {str(e)}"

print("ğŸ’± Tool de conversÃ£o de moedas criada!")
print("ğŸ¯ Agora vocÃª pode converter BRL, USD e EUR")
````
