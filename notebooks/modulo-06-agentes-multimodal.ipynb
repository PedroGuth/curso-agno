## ğŸ“¸ Trabalhando com Imagens no Agno

Vamos comeÃ§ar com o bÃ¡sico: **processamento de imagens**. No Agno 2.0, a classe `Image` foi completamente reformulada para ser mais simples e poderosa!

### Principais mudanÃ§as no Agno 2.0:

- âœ… **Classe `Image` unificada** (substitui `ImageArtifact`)
- âœ… **Armazenamento em bytes** (mais eficiente)
- âœ… **MÃ©todos padronizados** (`from_base64()`, `to_base64()`, etc)
- âœ… **IDs automÃ¡ticos** (UUID gerado automaticamente)
- âœ… **ValidaÃ§Ã£o automÃ¡tica** (sÃ³ uma fonte de conteÃºdo)

**Analogia:** Ã‰ como se antes vocÃª tinha 3 tipos diferentes de porta-retratos, agora tem um sÃ³ que serve para tudo! ğŸ–¼ï¸

````python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.media import Image
import requests
from PIL import Image as PILImage
import io
import base64

# Criando nosso agente multimodal
agente_visual = Agent(
    name="Agente Analisador Visual",
    model=Gemini(id="gemini-2.0-flash-exp"),  # Modelo gratuito do Google!
    description="Sou especialista em analisar imagens e extrair informaÃ§Ãµes detalhadas",
    instructions=[
        "Analise imagens de forma detalhada e didÃ¡tica",
        "Identifique objetos, pessoas, cenÃ¡rios e contextos",
        "Seja especÃ­fico e use linguagem brasileira"
    ],
    markdown=True
)

print("ğŸ¤– Agente Visual criado com sucesso!")
print(f"ğŸ“‹ Nome: {agente_visual.name}")
print(f"ğŸ§  Modelo: {agente_visual.model.id}")
````

````python
# Vamos criar uma imagem de exemplo simples
import matplotlib.pyplot as plt
import numpy as np

# Criando um grÃ¡fico simples para testar
fig, ax = plt.subplots(figsize=(10, 6))

# Dados de exemplo - vendas mensais
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [1200, 1400, 1100, 1600, 1800, 2000]

bars = ax.bar(meses, vendas, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD'])
ax.set_title('ğŸ“ˆ Vendas por MÃªs - Loja Virtual', fontsize=16, fontweight='bold')
ax.set_ylabel('Vendas (R$)', fontsize=12)
ax.set_xlabel('Meses', fontsize=12)

# Adicionando valores em cima das barras
for bar, valor in zip(bars, vendas):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20, 
            f'R$ {valor}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.grid(axis='y', alpha=0.3)

# Salvando a imagem em buffer
buffer = io.BytesIO()
plt.savefig(buffer, format='png', dpi=150, bbox_inches='tight')
buffer.seek(0)
image_bytes = buffer.getvalue()
buffer.close()

plt.show()

print("ğŸ“Š GrÃ¡fico criado com sucesso!")
print(f"ğŸ“ Tamanho da imagem: {len(image_bytes)} bytes")
````

````python
# Agora vamos criar um objeto Image do Agno
from agno.media import Image

# Criando o objeto Image com os bytes da imagem
imagem_grafico = Image(
    content=image_bytes,  # Os bytes da imagem
    name="grafico_vendas.png",
    caption="GrÃ¡fico de vendas mensais"
)

print("ğŸ–¼ï¸ Objeto Image criado!")
print(f"ğŸ†” ID: {imagem_grafico.id}")
print(f"ğŸ“ Nome: {imagem_grafico.name}")
print(f"ğŸ’¬ Caption: {imagem_grafico.caption}")
print(f"ğŸ“ Tamanho: {len(imagem_grafico.content)} bytes")

# Testando mÃ©todos Ãºteis
print(f"\nğŸ” MÃ©todos disponÃ­veis:")
print(f"ğŸ“¤ to_base64(): Converte para base64")
print(f"ğŸ“¥ from_base64(): Cria Image de base64")
print(f"ğŸ“Š to_dict(): Converte para dicionÃ¡rio")
print(f"ğŸ¯ get_content_bytes(): Pega bytes do conteÃºdo")
````

````python
# Agora vamos pedir para o agente analisar a imagem!
print("ğŸ” Enviando imagem para anÃ¡lise...")
print("â³ Aguarde, o agente estÃ¡ processando...\n")

# Fazendo a anÃ¡lise da imagem
resposta = agente_visual.run(
    input="Analise este grÃ¡fico detalhadamente. O que vocÃª consegue extrair de informaÃ§Ãµes?",
    images=[imagem_grafico]  # Passando a imagem para o agente
)

print("ğŸ¤– ANÃLISE DO AGENTE:")
print("=" * 50)
print(resposta.content)
print("=" * 50)

print(f"\nğŸ“Š MÃ©tricas da anÃ¡lise:")
print(f"ğŸ”¤ Tokens de entrada: {resposta.metrics.input_tokens}")
print(f"ğŸ”¤ Tokens de saÃ­da: {resposta.metrics.output_tokens}")
print(f"â±ï¸ Tempo de processamento: {resposta.metrics.duration:.2f}s")
````

## ğŸµ Trabalhando com Ãudio

Agora bora para o **Ã¡udio**! O Agno 2.0 tambÃ©m reformulou completamente como trabalhamos com arquivos de som.

### Nova Classe Audio:

- âœ… **Substitui** `AudioArtifact` e `AudioResponse`
- âœ… **Campos extras:** `transcript`, `sample_rate`, `channels`
- âœ… **Suporte completo** a diferentes formatos
- âœ… **IntegraÃ§Ã£o nativa** com modelos multimodais

**Analogia:** Ã‰ como ter um gravador profissional que nÃ£o sÃ³ grava, mas jÃ¡ vem com transcriÃ§Ã£o automÃ¡tica! ğŸ¤

**Dica!** Vamos simular um Ã¡udio usando texto-para-fala conceitual, jÃ¡ que nÃ£o temos acesso direto a APIs de Ã¡udio.

````python
from agno.media import Audio
import json

# Vamos simular um objeto Audio com dados conceituais
# Em um cenÃ¡rio real, vocÃª teria bytes de Ã¡udio de verdade

# Simulando metadados de um Ã¡udio
audio_simulado = Audio(
    name="reuniao_vendas.mp3",
    caption="GravaÃ§Ã£o da reuniÃ£o de vendas do trimestre",
    transcript="OlÃ¡ pessoal, vamos analisar os resultados do trimestre. As vendas de janeiro foram de 1200 reais, fevereiro 1400, marÃ§o 1100, abril 1600, maio 1800 e junho 2000. Houve um crescimento significativo no final do perÃ­odo.",
    sample_rate=44100,
    channels=2,
    # Para demonstraÃ§Ã£o, vamos usar dados simulados
    content=b"audio_data_simulado"  
)

print("ğŸµ Objeto Audio criado!")
print(f"ğŸ†” ID: {audio_simulado.id}")
print(f"ğŸ“ Nome: {audio_simulado.name}")
print(f"ğŸ“ Caption: {audio_simulado.caption}")
print(f"ğŸ™ï¸ Sample Rate: {audio_simulado.sample_rate} Hz")
print(f"ğŸ”Š Canais: {audio_simulado.channels}")
print(f"ğŸ“œ TranscriÃ§Ã£o: {audio_simulado.transcript[:100]}...")

print("\nâœ¨ Propriedades especiais do Audio:")
print("ğŸ¯ transcript: TranscriÃ§Ã£o automÃ¡tica do Ã¡udio")
print("ğŸ›ï¸ sample_rate: Taxa de amostragem (qualidade)")
print("ğŸ”Š channels: NÃºmero de canais (mono/stereo)")
print("â° expires_at: Data de expiraÃ§Ã£o (opcional)")
````

````python
# Criando um agente especializado em anÃ¡lise de Ã¡udio/transcriÃ§Ãµes
agente_audio = Agent(
    name="Agente Analisador de Ãudio",
    model=Gemini(id="gemini-2.0-flash-exp"),
    description="Especialista em analisar transcriÃ§Ãµes de Ã¡udio e extrair insights",
    instructions=[
        "Analise transcriÃ§Ãµes de Ã¡udio para extrair informaÃ§Ãµes importantes",
        "Identifique pontos-chave, tendÃªncias e insights",
        "Seja objetivo e destacue os dados mais relevantes",
        "Use linguagem brasileira e seja didÃ¡tico"
    ],
    markdown=True
)

print("ğŸ¤ Agente de Ãudio criado!")
print("ğŸš€ Vamos analisar a transcriÃ§Ã£o...")
````

````python
# Analisando a transcriÃ§Ã£o do Ã¡udio
print("ğŸ§ Enviando transcriÃ§Ã£o para anÃ¡lise...")
print("â³ Processando Ã¡udio...\n")

prompt_audio = f"""
Analise esta transcriÃ§Ã£o de uma reuniÃ£o de vendas:

**TranscriÃ§Ã£o:** {audio_simulado.transcript}

**Metadados do Ã¡udio:**
- Arquivo: {audio_simulado.name}
- Qualidade: {audio_simulado.sample_rate} Hz, {audio_simulado.channels} canais
- Contexto: {audio_simulado.caption}

Extraia insights, tendÃªncias e pontos importantes!
"""

resposta_audio = agente_audio.run(input=prompt_audio)

print("ğŸ¤ ANÃLISE DA TRANSCRIÃ‡ÃƒO:")
print("=" * 50)
print(resposta_audio.content)
print("=" * 50)

print(f"\nğŸ“Š MÃ©tricas da anÃ¡lise:")
print(f"â±ï¸ Tempo: {resposta_audio.metrics.duration:.2f}s")
print(f"ğŸ”¤ Tokens: {resposta_audio.metrics.input_tokens} â†’ {resposta_audio.metrics.output_tokens}")
````

## ğŸ¬ Processamento de VÃ­deo

Bora para o prÃ³ximo nÃ­vel: **vÃ­deos**! O Agno 2.0 tambÃ©m tem suporte completo para processamento de vÃ­deo.

### Classe Video Unificada:

- âœ… **Processamento temporal** (frames ao longo do tempo)
- âœ… **Metadados completos** (duraÃ§Ã£o, fps, resoluÃ§Ã£o)
- âœ… **AnÃ¡lise de conteÃºdo** visual e contextual
- âœ… **IntegraÃ§Ã£o perfeita** com modelos multimodais

**Analogia:** Ã‰ como ter um assistente que nÃ£o sÃ³ assiste ao filme com vocÃª, mas consegue pausar, voltar e analisar cada cena! ğŸ¥

**Dica!** VÃ­deos sÃ£o processados como sequÃªncias de frames, entÃ£o modelos conseguem "ver" a progressÃ£o temporal!

````python
from agno.media import Video
import datetime

# Simulando um objeto Video (em cenÃ¡rio real teria bytes de vÃ­deo)
video_exemplo = Video(
    name="apresentacao_resultados.mp4",
    caption="VÃ­deo da apresentaÃ§Ã£o dos resultados trimestrais",
    # Em um caso real, aqui estariam os bytes do vÃ­deo
    content=b"video_data_simulado",
    # Metadados Ãºteis para anÃ¡lise
    duration=180,  # 3 minutos
    fps=30,
    resolution="1920x1080"
)

print("ğŸ¬ Objeto Video criado!")
print(f"ğŸ†” ID: {video_exemplo.id}")
print(f"ğŸ“ Nome: {video_exemplo.name}")
print(f"ğŸ“ DescriÃ§Ã£o: {video_exemplo.caption}")
print(f"â±ï¸ DuraÃ§Ã£o: {video_exemplo.duration} segundos")
print(f"ğŸï¸ FPS: {video_exemplo.fps}")
print(f"ğŸ“º ResoluÃ§Ã£o: {video_exemplo.resolution}")

# Vamos simular alguns "frames" conceituais do vÃ­deo
frames_simulados = [
    "Frame 0:00 - Slide de abertura: 'Resultados Q2 2024'",
    "Frame 0:30 - GrÃ¡fico de barras mostrando vendas mensais",
    "Frame 1:00 - Apresentador explicando crescimento",
    "Frame 1:30 - Slide com metas para prÃ³ximo trimestre",
    "Frame 2:00 - GrÃ¡fico de pizza com distribuiÃ§Ã£o por produto",
    "Frame 2:30 - Slide de conclusÃ£o e perguntas"
]

print("\nğŸï¸ Frames simulados do vÃ­deo:")
for i, frame in enumerate(frames_simulados):
    print(f"  {i+1}. {frame}")
````

````python
# Criando agente especializado em anÃ¡lise de vÃ­deo
agente_video = Agent(
    name="Agente Analisador de VÃ­deo",
    model=Gemini(id="gemini-2.0-flash-exp"),
    description="Especialista em analisar conteÃºdo de vÃ­deos e extrair informaÃ§Ãµes temporais",
    instructions=[
        "Analise vÃ­deos considerando a progressÃ£o temporal do conteÃºdo",
        "Identifique momentos-chave, transiÃ§Ãµes e narrativa",
        "Correlacione informaÃ§Ãµes visuais com contexto temporal",
        "Seja detalhado sobre a estrutura do vÃ­deo"
    ],
    markdown=True
)

# Preparando descriÃ§Ã£o detalhada do vÃ­deo
descricao_video = f"""
**AnÃ¡lise de VÃ­deo de ApresentaÃ§Ã£o**

**Metadados:**
- Nome: {video_exemplo.name}
- DuraÃ§Ã£o: {video_exemplo.duration} segundos ({video_exemplo.duration//60}:{video_exemplo.duration%60:02d})
- Qualidade: {video_exemplo.resolution} @ {video_exemplo.fps} FPS
- Contexto: {video_exemplo.caption}

**SequÃªncia Temporal dos Frames:**
"""

for frame in frames_simulados:
    descricao_video += f"\n- {frame}"

descricao_video += """

Analise a estrutura narrativa, progressÃ£o do conteÃºdo e insights que podem ser extraÃ­dos desta apresentaÃ§Ã£o!
"""

print("ğŸ¥ Enviando vÃ­deo para anÃ¡lise...")
print("â³ Analisando sequÃªncia temporal...\n")

resposta_video = agente_video.run(input=descricao_video)

print("ğŸ¬ ANÃLISE DO VÃDEO:")
print("=" * 50)
print(resposta_video.content)
print("=" * 50)
````

## ğŸ“„ Trabalhando com Arquivos (Files)

Agora vamos para **arquivos**! PDFs, documentos, planilhas - tudo que nÃ£o Ã© mÃ­dia audiovisual.

````python
from agno.media import File
import json

# Vamos simular um arquivo CSV com dados de vendas
conteudo_csv = """mes,vendas,regiao,produto
Janeiro,1200,Sudeste,Produto A
Janeiro,800,Sul,Produto B
Fevereiro,1400,Sudeste,Produto A
Fevereiro,900,Sul,Produto B
Marco,1100,Sudeste,Produto A
Marco,750,Sul,Produto B
Abril,1600,Sudeste,Produto A
Abril,1050,Sul,Produto B
Maio,1800,Sudeste,Produto A
Maio,1200,Sul,Produto B
Junho,2000,Sudeste,Produto A
Junho,1300,Sul,Produto B"""

# Criando objeto File
arquivo_vendas = File(
    name="vendas_trimestre.csv",
    content=conteudo_csv.encode('utf-8'),  # Convertendo para bytes
    content_type="text/csv",
    caption="Dados de vendas por mÃªs, regiÃ£o e produto",
    size=len(conteudo_csv.encode('utf-8'))
)

print("ğŸ“„ Objeto File criado!")
print(f"ğŸ†” ID: {arquivo_vendas.id}")
print(f"ğŸ“ Nome: {arquivo_vendas.name}")
print(f"ğŸ“Š Tipo: {arquivo_vendas.content_type}")
print(f"ğŸ“ Tamanho: {arquivo_vendas.size} bytes")
print(f"ğŸ’¬ DescriÃ§Ã£o: {arquivo_vendas.caption}")

print("\nğŸ“‹ PrÃ©via do conteÃºdo:")
print(conteudo_csv[:200] + "...")

print("\nâœ¨ Vantagens da classe File:")
print("ğŸ” ExtraÃ§Ã£o automÃ¡tica de conteÃºdo")
print("ğŸ“Š DetecÃ§Ã£o de tipo de arquivo")
print("ğŸ§  IntegraÃ§Ã£o com Knowledge Base")
print("ğŸ¯ Processamento inteligente por contexto")
````

````python
# Criando agente especializado em anÃ¡lise de arquivos
agente_arquivos = Agent(
    name="Agente Analisador de Arquivos",
    model=Gemini(id="gemini-2.0-flash-exp"),
    description="Especialista em processar e analisar conteÃºdo de arquivos estruturados",
    instructions=[
        "Analise arquivos identificando padrÃµes e estruturas",
        "Para CSVs, identifique colunas, tipos de dados e tendÃªncias",
        "Extraia insights estatÃ­sticos e sugestÃµes de anÃ¡lise",
        "Seja especÃ­fico sobre a estrutura dos dados"
    ],
    markdown=True
)

# Preparando anÃ¡lise do arquivo
prompt_arquivo = f"""
Analise este arquivo CSV de vendas:

**InformaÃ§Ãµes do Arquivo:**
- Nome: {arquivo_vendas.name}
- Tipo: {arquivo_vendas.content_type}
- Tamanho: {arquivo_vendas.size} bytes
- DescriÃ§Ã£o: {arquivo_vendas.caption}

**ConteÃºdo:**
```csv
{conteudo_csv}
```

FaÃ§a uma anÃ¡lise completa: estrutura, padrÃµes, tendÃªncias e insights!
"""

print("ğŸ“Š Enviando arquivo para anÃ¡lise...")
print("â³ Processando dados estruturados...\n")

resposta_arquivo = agente_arquivos.run(input=prompt_arquivo)

print("ğŸ“„ ANÃLISE DO ARQUIVO:")
print("=" * 50)
print(resposta_arquivo.content)
print("=" * 50)
````

## ğŸ› ï¸ ParÃ¢metros de Controle de MÃ­dia

O Agno 2.0 trouxe controles avanÃ§ados para mÃ­dia. Vamos entender os **parÃ¢metros importantes**:

### Principais ParÃ¢metros:

- ğŸ“¤ **`send_media_to_model`**: Controla se mÃ­dia vai para o modelo

````python
# Demonstrando diferentes configuraÃ§Ãµes de mÃ­dia

# CONFIGURAÃ‡ÃƒO 1: MÃ­dia para modelo E armazenamento (padrÃ£o)
agente_completo = Agent(
    name="Agente Completo",
    model=Gemini(id="gemini-2.0-flash-exp"),
    send_media_to_model=True,  # MÃ­dia vai para o modelo
    store_media=True,          # MÃ­dia fica salva
    description="Processamento completo de mÃ­dia"
)

# CONFIGURAÃ‡ÃƒO 2: MÃ­dia sÃ³ para ferramentas
agente_tools_only = Agent(
    name="Agente Tools Only",
    model=Gemini(id="gemini-2.0-flash-exp"),
    send_media_to_model=False,  # MÃ­dia NÃƒO vai para o modelo
    store_media=True,           # Mas fica salva
    description="MÃ­dia processada apenas por ferramentas"
)

# CONFIGURAÃ‡ÃƒO 3: AnÃ¡lise rÃ¡pida (nÃ£o persistir)
agente_rapido = Agent(
    name="Agente RÃ¡pido",
    model=Gemini(id="gemini-2.0-flash-exp"),
    send_media_to_model=True,   # MÃ­dia vai para o modelo
    store_media=False,          # Mas NÃƒO fica salva
    description="AnÃ¡lise rÃ¡pida sem persistÃªncia"
)

print("ğŸ›ï¸ CONFIGURAÃ‡Ã•ES DE MÃDIA DEMONSTRADAS:")
print("\n1ï¸âƒ£ Agente Completo:")
print(f"   ğŸ“¤ Enviar para modelo: {agente_completo.send_media_to_model}")
print(f"   ğŸ’¾ Armazenar mÃ­dia: {agente_completo.store_media}")
print(f"   ğŸ¯ Uso: AnÃ¡lise completa com persistÃªncia")

print("\n2ï¸âƒ£ Agente Tools Only:")
print(f"   ğŸ“¤ Enviar para modelo: {agente_tools_only.send_media_to_model}")
print(f"   ğŸ’¾ Armazenar mÃ­dia: {agente_tools_only.store_media}")
print(f"   ğŸ¯ Uso: Processamento via ferramentas especializadas")

print("\n3ï¸âƒ£ Agente RÃ¡pido:")
print(f"   ğŸ“¤ Enviar para modelo: {agente_rapido.send_media_to_model}")
print(f"   ğŸ’¾ Armazenar mÃ­dia: {agente_rapido.store_media}")
print(f"   ğŸ¯ Uso: AnÃ¡lise rÃ¡pida sem overhead de armazenamento")

print("\nâœ¨ Escolha a configuraÃ§Ã£o ideal para seu caso de uso!")
````

## ğŸ”§ Ferramentas Multimodais AvanÃ§adas

Vamos criar uma **ferramenta customizada** que trabalha com mÃºltiplos tipos de mÃ­dia. Lembra do MÃ³dulo 3 sobre Tools? Agora vamos turbinar com multimodalidade!

### ParÃ¢metros Especiais para Tools:

- ğŸ“¸ **`images`**: Lista de imagens disponÃ­veis
- ğŸµ **`audio`**: Lista de Ã¡udios disponÃ­veis  
- ğŸ¬ **`videos`**: Lista de vÃ­deos disponÃ­veis
- ğŸ“„ **`files`**: Lista de arquivos disponÃ­veis

**ConexÃ£o com MÃ³dulo 3:** Lembra das ferramentas customizadas? Agora elas podem acessar mÃ­dia automaticamente!

**Dica!** Essas mÃ­dias sÃ£o injetadas automaticamente pelo Agno - vocÃª nÃ£o precisa passar manualmente!

````python
from agno.tools import tool
from typing import List, Optional
from agno.media import Image, Audio, Video, File

# Tool multimodal que processa diferentes tipos de mÃ­dia
@tool
def analisador_multimodal(
    tipo_analise: str,
    images: Optional[List[Image]] = None,
    audio: Optional[List[Audio]] = None, 
    videos: Optional[List[Video]] = None,
    files: Optional[List[File]] = None
) -> str:
    """
    Analisa diferentes tipos de mÃ­dia de forma integrada.
    
    Args:
        tipo_analise: Tipo de anÃ¡lise desejada (resumo, detalhada, comparativa)
        images: Lista de imagens (injetada automaticamente)
        audio: Lista de Ã¡udios (injetada automaticamente)
        videos: Lista de vÃ­deos (injetada automaticamente)
        files: Lista de arquivos (injetada automaticamente)
    """
    
    resultado = f"ğŸ¯ ANÃLISE MULTIMODAL - {tipo_analise.upper()}\n"
    resultado += "=" * 50 + "\n"
    
    # Contabilizando mÃ­dias disponÃ­veis
    total_midias = 0
    
    if images:
        total_midias += len(images)
        resultado += f"ğŸ“¸ IMAGENS ({len(images)}):" + "\n"
        for i, img in enumerate(images, 1):
            resultado += f"  {i}. {img.name or 'Imagem sem nome'}" + "\n"
            if img.caption:
                resultado += f"     ğŸ’¬ {img.caption}" + "\n"
            resultado += f"     ğŸ“ Tamanho: {len(img.content) if img.content else 0} bytes" + "\n"
    
    if audio:
        total_midias += len(audio)
        resultado += f"\nğŸµ ÃUDIOS ({len(audio)}):" + "\n"
        for i, aud in enumerate(audio, 1):
            resultado += f"  {i}. {aud.name or 'Ãudio sem nome'}" + "\n"
            if aud.transcript:
                resultado += f"     ğŸ“œ TranscriÃ§Ã£o: {aud.transcript[:100]}..." + "\n"
            if aud.sample_rate:
                resultado += f"     ğŸ›ï¸ Sample Rate: {aud.sample_rate} Hz" + "\n"
    
    if videos:
        total_midias += len(videos)
        resultado += f"\nğŸ¬ VÃDEOS ({len(videos)}):" + "\n"
        for i, vid in enumerate(videos, 1):
            resultado += f"  {i}. {vid.name or 'VÃ­deo sem nome'}" + "\n"
            if hasattr(vid, 'duration') and vid.duration:
                resultado += f"     â±ï¸ DuraÃ§Ã£o: {vid.duration}s" + "\n"
    
    if files:
        total_midias += len(files)
        resultado += f"\nğŸ“„ ARQUIVOS ({len(files)}):" + "\n"
        for i, arquivo in enumerate(files, 1):
            resultado += f"  {i}. {arquivo.name or 'Arquivo sem nome'}" + "\n"
            if arquivo.content_type:
                resultado += f"     ğŸ·ï¸ Tipo: {arquivo.content_type}" + "\n"
            if arquivo.size:
                resultado += f"     ğŸ“ Tamanho: {arquivo.size} bytes" + "\n"
    
    resultado += f"\nğŸ¯ RESUMO:" + "\n"
    resultado += f"ğŸ“Š Total de mÃ­dias processadas: {total_midias}" + "\n"
    resultado += f"ğŸ” Tipo de anÃ¡lise: {tipo_analise}" + "\n"
    
    if total_midias == 0:
        resultado += "âš ï¸ Nenhuma mÃ­dia foi fornecida para anÃ¡lise!" + "\n"
    else:
        resultado += "âœ… AnÃ¡lise multimodal concluÃ­da com sucesso!" + "\n"
    
    return resultado

print("ğŸ”§ Ferramenta multimodal customizada criada!")
print("âœ¨ Funcionalidades:")
print("  ğŸ“¸ Processa imagens automaticamente")
print("  ğŸµ Analisa Ã¡udios disponÃ­veis")
print("  ğŸ¬ Examina vÃ­deos fornecidos")
print("  ğŸ“„ Processa arquivos estruturados")
print("  ğŸ§  Integra todos os dados em anÃ¡lise Ãºnica")
````

````python
# Criando agente com a ferramenta multimodal customizada
agente_com_tool = Agent(
    name="Agente com Tool Multimodal",
    model=Gemini(id="gemini-2.0-flash-exp"),
    tools=[analisador_multimodal],  # Nossa tool customizada!
    description="Agente equipado com ferramenta de anÃ¡lise multimodal avanÃ§ada",
    instructions=[
        "Use a ferramenta analisador_multimodal para processar mÃ­dias",
        "Combine os resultados da ferramenta com sua anÃ¡lise prÃ³pria",
        "Gere insights Ãºnicos baseados na anÃ¡lise integrada"
    ],
    markdown=True
)

print("ğŸ¤– Agente com ferramenta multimodal criado!")
print("ğŸ”§ Ferramenta customizada ativa: analisador_multimodal")
print("ğŸš€ Pronto para anÃ¡lise avanÃ§ada!")
````
