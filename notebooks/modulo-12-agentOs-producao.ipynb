## üèóÔ∏è Criando Nosso Primeiro AgentOS

Agora vamos criar um agente e transform√°-lo numa API de produ√ß√£o usando o AgentOS!

√â tipo transformar um carro comum num Uber - mesma funcionalidade, mas agora qualquer um pode usar! üöó‚û°Ô∏èüöï

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.db.sqlite import SqliteDb
from agno.os import AgentOS
from agno.tools.duckduckgo import DuckDuckGoTools

# Vamos criar um agente especialista em tecnologia
tech_agent = Agent(
    name="TechExpert",
    model=Gemini(id="gemini-2.0-flash"),
    description="Sou um especialista em tecnologia que ajuda com d√∫vidas sobre programa√ß√£o, IA, e tend√™ncias tech.",
    instructions=[
        "Sempre forne√ßa respostas t√©cnicas precisas",
        "Use exemplos pr√°ticos quando poss√≠vel",
        "Seja did√°tico e explique conceitos complexos de forma simples",
        "Quando n√£o souber algo, pesquise na web para dar informa√ß√µes atualizadas"
    ],
    tools=[DuckDuckGoTools()],
    db=SqliteDb(db_file="tech_agent.db"),
    add_history_to_context=True,
    markdown=True
)

print("ü§ñ Agente TechExpert criado!")
print("üìä Com banco de dados para persistir conversas")
print("üîç Com ferramentas de busca na web")
```

```python
# Agora vamos criar o AgentOS - o cora√ß√£o da produ√ß√£o!
agent_os = AgentOS(
    agents=[tech_agent],
    name="TechOS",
    description="Sistema de produ√ß√£o para o agente especialista em tecnologia"
)

# Pegamos a aplica√ß√£o FastAPI pronta!
app = agent_os.get_app()

print("üöÄ AgentOS criado com sucesso!")
print("‚ö° FastAPI app pronta para produ√ß√£o!")
print("üì° Endpoints autom√°ticos criados!")
```

## üîê Seguran√ßa e Boas Pr√°ticas

Agora que temos nosso agente rodando, precisamos deix√°-lo seguro! √â tipo trancar a porta de casa - b√°sico mas essencial! üè†üîê

### üõ°Ô∏è Checklist de Seguran√ßa:

1. **Vari√°veis de Ambiente** ‚úÖ
2. **Rate Limiting** ‚úÖ  
3. **Valida√ß√£o de Input** ‚úÖ
4. **Logs de Auditoria** ‚úÖ
5. **HTTPS** ‚úÖ

```python
# Vamos criar um setup de seguran√ßa avan√ßado
security_config = '''
import os
from fastapi import FastAPI, HTTPException, Depends, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.middleware.trustedhost import TrustedHostMiddleware
from slowapi import Limiter, _rate_limit_exceeded_handler
from slowapi.util import get_remote_address
from slowapi.errors import RateLimitExceeded
import logging
from datetime import datetime

# Configurar logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('agent_security.log'),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger("AgentOS")

# Rate limiting
limiter = Limiter(key_func=get_remote_address)

def create_secure_app(agent_os):
    app = agent_os.get_app()

    # CORS
    app.add_middleware(
        CORSMiddleware,
        allow_origins=["https://yourdomain.com"],  # Apenas dom√≠nios confi√°veis
        allow_credentials=True,
        allow_methods=["POST", "GET"],
        allow_headers=["*"]
    )

    # Trusted hosts
    app.add_middleware(
        TrustedHostMiddleware,
        allowed_hosts=["yourdomain.com", "*.yourdomain.com"]
    )

    # Rate limiting
    app.state.limiter = limiter
    app.add_exception_handler(RateLimitExceeded, _rate_limit_exceeded_handler)

    # Middleware de auditoria
    @app.middleware("http")
    async def audit_middleware(request: Request, call_next):
        start_time = datetime.now()

        # Log da requisi√ß√£o
        logger.info(f"Request: {request.method} {request.url} from {request.client.host}")

        response = await call_next(request)

        # Log da resposta
        process_time = (datetime.now() - start_time).total_seconds()
        logger.info(f"Response: {response.status_code} in {process_time:.2f}s")

        return response

    return app
'''

with open('security_config.py', 'w') as f:
    f.write(security_config)

print("üîê Configura√ß√£o de seguran√ßa criada!")
print("üìù Arquivo: security_config.py")
print("\nüõ°Ô∏è Recursos de seguran√ßa inclu√≠dos:")
print("  ‚úÖ Rate Limiting (prote√ß√£o contra spam)")
print("  ‚úÖ CORS configurado")
print("  ‚úÖ Trusted hosts")
print("  ‚úÖ Logs de auditoria")
print("  ‚úÖ Middleware de monitoramento")
```

```python
# Configura√ß√£o completa de produ√ß√£o
production_config = '''
import os
import logging
from agno.agent import Agent
from agno.team import Team
from agno.models.google import Gemini
from agno.db.postgres import PostgresDb
from agno.os import AgentOS
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.hackernews import HackerNewsTools
import uvicorn

# Configura√ß√£o de logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('production.log'),
        logging.StreamHandler()
    ]
)

def create_production_system():
    """Cria sistema completo de produ√ß√£o"""

    # Database de produ√ß√£o (PostgreSQL)
    db = PostgresDb(
        db_url=os.getenv("DATABASE_URL", "postgresql://user:pass@localhost/agno"),
        session_table="production_sessions",
        memory_table="production_memories"
    )

    # Agente especialista em tech
    tech_agent = Agent(
        name="TechExpert",
        model=Gemini(
            id="gemini-2.0-flash",
            temperature=0.3,  # Mais conservador em produ√ß√£o
            max_tokens=2048
        ),
        description="Especialista em tecnologia, programa√ß√£o e IA",
        instructions=[
            "Sempre forne√ßa respostas t√©cnicas precisas",
            "Use fontes confi√°veis e atualizadas",
            "Seja did√°tico mas direto ao ponto",
            "Em caso de d√∫vida, pesquise informa√ß√µes atuais"
        ],
        tools=[DuckDuckGoTools(), HackerNewsTools()],
        db=db,
        add_history_to_context=True,
        num_history_runs=5,
        enable_user_memories=True,
        markdown=True
    )

    # Agente de suporte
    support_agent = Agent(
        name="SupportBot",
        model=Gemini(id="gemini-2.0-flash"),
        description="Agente de suporte ao cliente especializado em resolver d√∫vidas",
        instructions=[
            "Seja sempre educado e prestativo",
            "Resolva problemas de forma eficiente",
            "Escale para humanos quando necess√°rio"
        ],
        db=db,
        add_history_to_context=True,
        enable_user_memories=True
    )

    # Time de agentes
    agent_team = Team(
        name="TechTeam",
        members=[tech_agent, support_agent],
        instructions="Trabalhem juntos para resolver qualquer quest√£o do usu√°rio",
        db=db
    )

    # AgentOS
    agent_os = AgentOS(
        agents=[tech_agent, support_agent],
        teams=[agent_team],
        name="ProductionOS",
        description="Sistema de produ√ß√£o completo"
    )

    return agent_os

def main():
    """Fun√ß√£o principal"""

    print("üöÄ Iniciando sistema de produ√ß√£o...")

    # Criar sistema
    agent_os = create_production_system()
    app = agent_os.get_app()

    # Configura√ß√µes de produ√ß√£o
    config = {
        "host": "0.0.0.0",
        "port": int(os.getenv("PORT", 8000)),
        "workers": int(os.getenv("WORKERS", 4)),
        "log_level": "info",
        "access_log": True
    }

    print(f"üåê Servidor iniciando em {config['host']}:{config['port']}")
    print(f"üë• Usando {config['workers']} workers")

    # Iniciar servidor
    uvicorn.run(app, **config)

if __name__ == "__main__":
    main()
'''

with open('production_system.py', 'w') as f:
    f.write(production_config)

print("üè≠ Sistema de produ√ß√£o completo criado!")
print("üìù Arquivo: production_system.py")
print("\nüîß Features inclu√≠das:")
print("  ‚úÖ Multiple agents & teams")
print("  ‚úÖ PostgreSQL database")
print("  ‚úÖ Logging completo")
print("  ‚úÖ Configura√ß√£o por vari√°veis de ambiente")
print("  ‚úÖ Multiple workers")
print("  ‚úÖ Mem√≥ria persistente")
```
