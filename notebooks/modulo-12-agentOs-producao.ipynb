## ğŸ—ï¸ Criando Nosso Primeiro AgentOS

Agora vamos criar um agente e transformÃ¡-lo numa API de produÃ§Ã£o usando o AgentOS!

Ã‰ tipo transformar um carro comum num Uber - mesma funcionalidade, mas agora qualquer um pode usar! ğŸš—â¡ï¸ğŸš•

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.db.sqlite import SqliteDb
from agno.os import AgentOS
from agno.tools.duckduckgo import DuckDuckGoTools

# Vamos criar um agente especialista em tecnologia
tech_agent = Agent(
    name="TechExpert",
    model=Gemini(id="gemini-2.0-flash"),
    description="Sou um especialista em tecnologia que ajuda com dÃºvidas sobre programaÃ§Ã£o, IA, e tendÃªncias tech.",
    instructions=[
        "Sempre forneÃ§a respostas tÃ©cnicas precisas",
        "Use exemplos prÃ¡ticos quando possÃ­vel",
        "Seja didÃ¡tico e explique conceitos complexos de forma simples",
        "Quando nÃ£o souber algo, pesquise na web para dar informaÃ§Ãµes atualizadas"
    ],
    tools=[DuckDuckGoTools()],
    db=SqliteDb(db_file="tech_agent.db"),
    add_history_to_context=True,
    markdown=True
)

print("ğŸ¤– Agente TechExpert criado!")
print("ğŸ“Š Com banco de dados para persistir conversas")
print("ğŸ” Com ferramentas de busca na web")
```

```python
# Agora vamos criar o AgentOS - o coraÃ§Ã£o da produÃ§Ã£o!
agent_os = AgentOS(
    agents=[tech_agent],
    name="TechOS",
    description="Sistema de produÃ§Ã£o para o agente especialista em tecnologia"
)

# Pegamos a aplicaÃ§Ã£o FastAPI pronta!
app = agent_os.get_app()

print("ğŸš€ AgentOS criado com sucesso!")
print("âš¡ FastAPI app pronta para produÃ§Ã£o!")
print("ğŸ“¡ Endpoints automÃ¡ticos criados!")
```

## ğŸ“Š Monitoramento e MÃ©tricas

Uma das partes mais importantes da produÃ§Ã£o Ã© saber como seus agentes estÃ£o performando!

Ã‰ tipo ter um painel no carro - vocÃª precisa saber se tÃ¡ tudo ok! ğŸš—ğŸ“Š

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
import random

# Vamos simular dados de monitoramento
def generate_monitoring_data():
    """Gera dados simulados de monitoramento"""

    # Ãšltimas 24 horas
    hours = []
    requests_per_hour = []
    response_time = []
    success_rate = []

    for i in range(24):
        hour = datetime.now() - timedelta(hours=23-i)
        hours.append(hour.strftime('%H:%M'))

        # Simular dados realistas
        base_requests = 50 + random.randint(-20, 30)
        if 9 <= hour.hour <= 17:  # HorÃ¡rio comercial
            base_requests *= 2

        requests_per_hour.append(base_requests)
        response_time.append(random.uniform(0.5, 3.0))
        success_rate.append(random.uniform(95, 100))

    return hours, requests_per_hour, response_time, success_rate

# Gerar dados
hours, requests, response_times, success_rates = generate_monitoring_data()

# Criar dashboard de monitoramento
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('ğŸ“Š Dashboard de Monitoramento AgentOS', fontsize=16, fontweight='bold')

# GrÃ¡fico 1: RequisiÃ§Ãµes por hora
ax1.plot(range(len(hours)), requests, marker='o', linewidth=2, color='#3498db')
ax1.set_title('ğŸ“ˆ RequisiÃ§Ãµes por Hora', fontweight='bold')
ax1.set_ylabel('NÃºmero de RequisiÃ§Ãµes')
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# GrÃ¡fico 2: Tempo de resposta
ax2.plot(range(len(hours)), response_times, marker='s', linewidth=2, color='#e74c3c')
ax2.set_title('â±ï¸ Tempo de Resposta MÃ©dio', fontweight='bold')
ax2.set_ylabel('Tempo (segundos)')
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='x', rotation=45)

# GrÃ¡fico 3: Taxa de sucesso
ax3.plot(range(len(hours)), success_rates, marker='^', linewidth=2, color='#2ecc71')
ax3.set_title('âœ… Taxa de Sucesso (%)', fontweight='bold')
ax3.set_ylabel('Porcentagem')
ax3.set_ylim(90, 100)
ax3.grid(True, alpha=0.3)
ax3.tick_params(axis='x', rotation=45)

# GrÃ¡fico 4: DistribuiÃ§Ã£o de tokens
token_usage = {
    'Input Tokens': sum(requests) * 25,
    'Output Tokens': sum(requests) * 150,
    'Cache Hits': sum(requests) * 10
}

ax4.pie(token_usage.values(), labels=token_usage.keys(), autopct='%1.1f%%',
        colors=['#f39c12', '#9b59b6', '#1abc9c'])
ax4.set_title('ğŸ¯ DistribuiÃ§Ã£o de Tokens', fontweight='bold')

plt.tight_layout()
plt.show()

# MÃ©tricas em tempo real
total_requests = sum(requests)
avg_response_time = np.mean(response_times)
avg_success_rate = np.mean(success_rates)

print("\nğŸ”¥ MÃ‰TRICAS EM TEMPO REAL (Ãšltimas 24h)")
print("="*50)
print(f"ğŸ“Š Total de RequisiÃ§Ãµes: {total_requests:,}")
print(f"âš¡ Tempo MÃ©dio de Resposta: {avg_response_time:.2f}s")
print(f"âœ… Taxa de Sucesso: {avg_success_rate:.1f}%")
print(f"ğŸ’° Custo Estimado: ${(total_requests * 0.001):.2f}")
print(f"ğŸš€ Status: {'ğŸŸ¢ HEALTHY' if avg_success_rate > 95 else 'ğŸŸ¡ WARNING'}")
```

## ğŸ” SeguranÃ§a e Boas PrÃ¡ticas

Agora que temos nosso agente rodando, precisamos deixÃ¡-lo seguro! Ã‰ tipo trancar a porta de casa - bÃ¡sico mas essencial! ğŸ ğŸ”

### ğŸ›¡ï¸ Checklist de SeguranÃ§a:

1. **VariÃ¡veis de Ambiente** âœ…
2. **Rate Limiting** âœ…  
3. **ValidaÃ§Ã£o de Input** âœ…
4. **Logs de Auditoria** âœ…
5. **HTTPS** âœ…

```python
# Vamos criar um setup de seguranÃ§a avanÃ§ado
security_config = '''
import os
from fastapi import FastAPI, HTTPException, Depends, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.middleware.trustedhost import TrustedHostMiddleware
from slowapi import Limiter, _rate_limit_exceeded_handler
from slowapi.util import get_remote_address
from slowapi.errors import RateLimitExceeded
import logging
from datetime import datetime

# Configurar logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('agent_security.log'),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger("AgentOS")

# Rate limiting
limiter = Limiter(key_func=get_remote_address)

def create_secure_app(agent_os):
    app = agent_os.get_app()

    # CORS
    app.add_middleware(
        CORSMiddleware,
        allow_origins=["https://yourdomain.com"],  # Apenas domÃ­nios confiÃ¡veis
        allow_credentials=True,
        allow_methods=["POST", "GET"],
        allow_headers=["*"]
    )

    # Trusted hosts
    app.add_middleware(
        TrustedHostMiddleware,
        allowed_hosts=["yourdomain.com", "*.yourdomain.com"]
    )

    # Rate limiting
    app.state.limiter = limiter
    app.add_exception_handler(RateLimitExceeded, _rate_limit_exceeded_handler)

    # Middleware de auditoria
    @app.middleware("http")
    async def audit_middleware(request: Request, call_next):
        start_time = datetime.now()

        # Log da requisiÃ§Ã£o
        logger.info(f"Request: {request.method} {request.url} from {request.client.host}")

        response = await call_next(request)

        # Log da resposta
        process_time = (datetime.now() - start_time).total_seconds()
        logger.info(f"Response: {response.status_code} in {process_time:.2f}s")

        return response

    return app
'''

with open('security_config.py', 'w') as f:
    f.write(security_config)

print("ğŸ” ConfiguraÃ§Ã£o de seguranÃ§a criada!")
print("ğŸ“ Arquivo: security_config.py")
print("\nğŸ›¡ï¸ Recursos de seguranÃ§a incluÃ­dos:")
print("  âœ… Rate Limiting (proteÃ§Ã£o contra spam)")
print("  âœ… CORS configurado")
print("  âœ… Trusted hosts")
print("  âœ… Logs de auditoria")
print("  âœ… Middleware de monitoramento")
```

## ğŸ› ï¸ ConfiguraÃ§Ã£o AvanÃ§ada de ProduÃ§Ã£o

Agora vamos criar uma configuraÃ§Ã£o de produÃ§Ã£o robusta com tudo que aprendemos!

```python
# ConfiguraÃ§Ã£o completa de produÃ§Ã£o
production_config = '''
import os
import logging
from agno.agent import Agent
from agno.team import Team
from agno.models.google import Gemini
from agno.db.postgres import PostgresDb
from agno.os import AgentOS
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.hackernews import HackerNewsTools
import uvicorn

# ConfiguraÃ§Ã£o de logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('production.log'),
        logging.StreamHandler()
    ]
)

def create_production_system():
    """Cria sistema completo de produÃ§Ã£o"""

    # Database de produÃ§Ã£o (PostgreSQL)
    db = PostgresDb(
        db_url=os.getenv("DATABASE_URL", "postgresql://user:pass@localhost/agno"),
        session_table="production_sessions",
        memory_table="production_memories"
    )

    # Agente especialista em tech
    tech_agent = Agent(
        name="TechExpert",
        model=Gemini(
            id="gemini-2.0-flash",
            temperature=0.3,  # Mais conservador em produÃ§Ã£o
            max_tokens=2048
        ),
        description="Especialista em tecnologia, programaÃ§Ã£o e IA",
        instructions=[
            "Sempre forneÃ§a respostas tÃ©cnicas precisas",
            "Use fontes confiÃ¡veis e atualizadas",
            "Seja didÃ¡tico mas direto ao ponto",
            "Em caso de dÃºvida, pesquise informaÃ§Ãµes atuais"
        ],
        tools=[DuckDuckGoTools(), HackerNewsTools()],
        db=db,
        add_history_to_context=True,
        num_history_runs=5,
        enable_user_memories=True,
        markdown=True
    )

    # Agente de suporte
    support_agent = Agent(
        name="SupportBot",
        model=Gemini(id="gemini-2.0-flash"),
        description="Agente de suporte ao cliente especializado em resolver dÃºvidas",
        instructions=[
            "Seja sempre educado e prestativo",
            "Resolva problemas de forma eficiente",
            "Escale para humanos quando necessÃ¡rio"
        ],
        db=db,
        add_history_to_context=True,
        enable_user_memories=True
    )

    # Time de agentes
    agent_team = Team(
        name="TechTeam",
        members=[tech_agent, support_agent],
        instructions="Trabalhem juntos para resolver qualquer questÃ£o do usuÃ¡rio",
        db=db
    )

    # AgentOS
    agent_os = AgentOS(
        agents=[tech_agent, support_agent],
        teams=[agent_team],
        name="ProductionOS",
        description="Sistema de produÃ§Ã£o completo"
    )

    return agent_os

def main():
    """FunÃ§Ã£o principal"""

    print("ğŸš€ Iniciando sistema de produÃ§Ã£o...")

    # Criar sistema
    agent_os = create_production_system()
    app = agent_os.get_app()

    # ConfiguraÃ§Ãµes de produÃ§Ã£o
    config = {
        "host": "0.0.0.0",
        "port": int(os.getenv("PORT", 8000)),
        "workers": int(os.getenv("WORKERS", 4)),
        "log_level": "info",
        "access_log": True
    }

    print(f"ğŸŒ Servidor iniciando em {config['host']}:{config['port']}")
    print(f"ğŸ‘¥ Usando {config['workers']} workers")

    # Iniciar servidor
    uvicorn.run(app, **config)

if __name__ == "__main__":
    main()
'''

with open('production_system.py', 'w') as f:
    f.write(production_config)

print("ğŸ­ Sistema de produÃ§Ã£o completo criado!")
print("ğŸ“ Arquivo: production_system.py")
print("\nğŸ”§ Features incluÃ­das:")
print("  âœ… Multiple agents & teams")
print("  âœ… PostgreSQL database")
print("  âœ… Logging completo")
print("  âœ… ConfiguraÃ§Ã£o por variÃ¡veis de ambiente")
print("  âœ… Multiple workers")
print("  âœ… MemÃ³ria persistente")
```
